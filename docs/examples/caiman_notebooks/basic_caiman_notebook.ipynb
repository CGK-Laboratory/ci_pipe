```python
import caiman
from caiman.utils.utils import download_demo
```

```python
movie_path = download_demo('data_endoscope.tif')
```

```python
movie_path
```

```python
from caiman.source_extraction.cnmf import cnmf, params
parameter_dict = {
    'strides': (48, 48),
    'overlaps': (24, 24),
    'max_shifts': (6,6),
    'max_deviation_rigid': 3,
    'pw_rigid': True,
    'shifts_opencv': True,
    'border_nan': 'copy'
}

parameters = params.CNMFParams(params_dict=parameter_dict)
```

```python
from caiman.motion_correction import MotionCorrect
mc_handler = MotionCorrect(movie_path, **parameters.motion)
```

```python
mc_handler.motion_correct(save_movie=True);
```

```python
processed_files = mc_handler.mmap_file
```

```python
processed_files
```

```python
import shutil
import os
path_to_executed_file = processed_files[0]
destination_directory = "/home/nazarena/projects/fiuba/cgk/docs/examples/caiman_notebooks/output_dir"
destination_path = os.path.join(destination_directory, os.path.basename(path_to_executed_file))

shutil.move(path_to_executed_file, destination_path)
```
